```python
# Task 2: Mark the success/failed launches for each site on the map
# Example DataFrame for launch_sites with Latitude, Longitude, and Launch Success
launch_sites = pd.DataFrame({
    'Launch Site': ['CCAFS LC-40', 'VAFB SLC-4E', 'KSC LC-39A'],
    'Lat': [28.5623, 34.6321, 28.5733],
    'Long': [-80.5774, -120.6107, -80.6469],
    'Launch Success': ['Success', 'Failure', 'Success']  # Example launch outcomes
})

# Initialize the map centered at NASA Johnson Space Center
nasa_coordinate = [29.559684, -95.083097]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Loop through the DataFrame and add different markers for successful and failed launches
for index, site in launch_sites.iterrows():
    # Get the coordinates, launch site name, and outcome (success/failure)
    coordinate = [site['Lat'], site['Long']]
    launch_site_name = site['Launch Site']
    outcome = site['Launch Success']
    
    # Determine the color based on success or failure
    color = 'green' if outcome == 'Success' else 'red'
    
    # Create a circle for each launch site with the outcome color
    circle = folium.Circle(coordinate, radius=1000, color=color, fill=True).add_child(folium.Popup(f'{launch_site_name}: {outcome}'))
    
    # Create a marker with a text label for the site name
    marker = folium.map.Marker(
        coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12; color:{color};"><b>{launch_site_name}</b></div>',
        )
    )
    
    # Add the circle and marker to the map
    site_map.add_child(circle)
    site_map.add_child(marker)

# Display the map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
# Function to calculate distance between two coordinates
# Example spacex_df DataFrame (with launch site coordinates)
def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth's radius in kilometers

    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    
    distance = R * c
    return distance

def calculate_distances(spacex_df, proximities_df):
    distances = []
    
    for _, launch_site in spacex_df.iterrows():
        site_distances = []
        for _, proximity in proximities_df.iterrows():
            distance = haversine_distance(
                launch_site['Lat'], launch_site['Long'],
                proximity['Latitude'], proximity['Longitude']
            )
            site_distances.append(distance)
        distances.append(site_distances)
    
    distance_df = pd.DataFrame(
        distances, 
        index=spacex_df['Launch Site'],
        columns=proximities_df['Name']
    )
    
    return distance_df
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
